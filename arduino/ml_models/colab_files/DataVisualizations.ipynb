# Imports and Connect to Google Drive

```python
import pandas as pd
import plotly.express as px
import os
import plotly.graph_objects as go
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

# Set Variables For Graphing

```python
activity = 'seizure'
window_size = 10
readings_per_second = 20
path = '/content/drive/MyDrive/data/03-31-2021_snapshot/%s/' % activity

# only need if you want to graph just part of the data
time_range = [1200, 4200]
plot_range = [-50, 50]
```

# Visualization of Accel and Gyro

```python
window = int(window_size) * readings_per_second
    
for file_name in os.listdir(path):
    if file_name.endswith('csv'):
      url = '/content/drive/MyDrive/data/03-31-2021_snapshot/%s/%s' % (activity, file_name)
  
      df = pd.read_csv(url) 

      # gyro
      fig = go.Figure()
      fig.update_xaxes(title_text='Time')
      fig.update_yaxes(title_text='Change in Accel & Gyro')
      fig.add_trace(go.Scatter(x=df.time, y=df['g_x'],
                    mode='lines',
                    name='g_x'))
      fig.add_trace(go.Scatter(x=df.time, y=df['g_y'],
                    mode='lines',
                    name='g_y'))
      fig.add_trace(go.Scatter(x=df.time, y=df['g_z'],
                    mode='lines',
                    name='g_z'))
      
      # accel
      fig.add_trace(go.Scatter(x=df.time, y=df['a_x'],
                    mode='lines',
                    name='a_x'))
      fig.add_trace(go.Scatter(x=df.time, y=df['a_y'],
                    mode='lines',
                    name='a_y'))
      fig.add_trace(go.Scatter(x=df.time, y=df['a_z'],
                    mode='lines',
                    name='a_z'))

      print(file_name)     
      fig.show()
```

# Rolling Mean Visualization For Acceleration

```python
paths = ["car", "run", "leisure", "play", "seizure", "sleep", "walk"]

window = int(window_size) * readings_per_second

for file_name in os.listdir(path):
  if file_name.endswith('csv'):

      df = pd.read_csv('/content/drive/MyDrive/data/03-31-2021_snapshot/%s/%s' % (activity, file_name)) 

      fig = px.line(df.rolling(window, win_type='boxcar').mean(),y = ['a_x', 'a_y', 'a_z']) 
      fig.update_xaxes(title_text='Time', range=time_range)
      fig.update_yaxes(title_text='Change in Acceleration', range=plot_range)
        
      fig.show()  
```
